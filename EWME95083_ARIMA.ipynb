```python
for x in range(len(Biscuit_Price.columns)-1):
    x=x+1
#     print('x:-', x)
    for y in range(12):
#         print('y:-', y)
        date = Biscuit_Price.columns[x] +'-'+ Biscuit_Price['Unnamed: 0'][y]
        production = Biscuit_Production[Biscuit_Price.columns[x]][y]
        price= Biscuit_Price[Biscuit_Price.columns[x]][y]
#         print(month)
#         print(production)
#         print(price)

        data_dic={
            'date':date,
            'production':production,
            'price':price
        }
        data =data.append(pd.DataFrame(data=data_dic, index=[0]),ignore_index=True)
```

**ADCF test**

```python
from statsmodels.tsa.stattools import adfuller

adf_test = adfuller(data['price'])

print('AD_fuller statistic :- ', adf_test[0])
print('P Value :- ', adf_test[1])
```

```python
#log transformation

data['price_log'] = np.log(data['price'])
```

```python
data['price_log'].dropna().plot()
```

```python
data['price_log_diff']=data['price_log'].dropna()- data['price_log'].dropna().shift(1)
```

```python
adf_test = adfuller(data['price_log_diff'].dropna())

print('AD_fuller statistic :- ', adf_test[0])
print('P Value :- ', adf_test[1])
```

**Hyperparameter**

```python
import itertools
```

```python
p = range(0,4)
q = range(0,4)
d =range(0,3)
```

```python
pdq_ = list(itertools.product(p,d,q))
```

```python
len(pdq_)
```

```python
rmse_=[]
order_ =[]
```

```python
import warnings
warnings.filterwarnings('ignore')


for pdq in pdq_:
    try:
        model_ = ARIMA(train_data['price_log_diff'], exog=train_data['production'], order=pdq).fit()
        pred_ = model_.predict(start=len(train_data['price_log_diff']), end=(len(data)-2), exog=test_data['production'])
        pred_.index = pred_.index.strftime('%Y-%m')
        error_ = np.sqrt(mean_squared_error(test_data['price_log_diff'],pred_))
        order_.append(pdq)
        rmse_.append(error_)
        
        
    except:
        continue
```

```python
best_param = pd.DataFrame(index= order_, data=rmse_, columns=['RMSE'])
```

```python
best_param.sort_values(by=['RMSE'])
```

```python
#best param fro model 
best_param = best_param.sort_values(by=['RMSE'])
best_param['RMSE'].index[0]
```
